# Autoresearch Results Analysis
Load `output/autoresearch_results.csv` and visualize experiment results.
Run locally — no GPU needed.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

csv_path = os.path.join(os.path.dirname(os.path.abspath('')), 'output', 'autoresearch_results.csv')
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    print(f'Loaded {len(df)} experiments')
    display(df.head())
else:
    print(f'No results file found at {csv_path}')
    print('Run autoresearch first, then copy the CSV here.')
    df = pd.DataFrame()
```

## Best config per stage

```python
if not df.empty:
    best = df.loc[df.groupby(['game', 'stage'])['sr'].idxmax()]
    print(best[['game', 'stage', 'model', 'temperature', 'prompt_variant', 'sr', 'steps', 'status']].to_string(index=False))
else:
    print('No data')
```

## Model comparison

```python
if not df.empty:
    fig, ax = plt.subplots(figsize=(10, 5))
    df.groupby('model')['sr'].mean().sort_values().plot(kind='barh', ax=ax)
    ax.set_xlabel('Mean SR')
    ax.set_title('Model Comparison (Mean SR)')
    plt.tight_layout()
    plt.show()
else:
    print('No data')
```

## Temperature vs SR

```python
if not df.empty:
    fig, ax = plt.subplots(figsize=(10, 5))
    for model, group in df.groupby('model'):
        means = group.groupby('temperature')['sr'].mean()
        ax.plot(means.index, means.values, marker='o', label=model)
    ax.set_xlabel('Temperature')
    ax.set_ylabel('Mean SR')
    ax.set_title('Temperature vs SR by Model')
    ax.legend()
    plt.tight_layout()
    plt.show()
else:
    print('No data')
```

## Prompt variant comparison

```python
if not df.empty:
    fig, ax = plt.subplots(figsize=(8, 4))
    df.groupby('prompt_variant')['sr'].mean().sort_values().plot(kind='barh', ax=ax)
    ax.set_xlabel('Mean SR')
    ax.set_title('Prompt Variant Comparison')
    plt.tight_layout()
    plt.show()
else:
    print('No data')
```

## Stage progression over time

```python
if not df.empty:
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(df['experiment_id'], df['sr'], marker='.', alpha=0.5)
    ax.set_xlabel('Experiment #')
    ax.set_ylabel('SR')
    ax.set_title('SR over experiments')
    plt.tight_layout()
    plt.show()
else:
    print('No data')
```

## SR vs Steps (Pareto front)

```python
if not df.empty:
    solved = df[df['sr'] > 0]
    if not solved.empty:
        fig, ax = plt.subplots(figsize=(8, 6))
        for model, group in solved.groupby('model'):
            ax.scatter(group['steps'], group['sr'], label=model, alpha=0.6)
        ax.set_xlabel('Steps')
        ax.set_ylabel('SR')
        ax.set_title('SR vs Steps')
        ax.legend()
        plt.tight_layout()
        plt.show()
    else:
        print('No solved experiments')
else:
    print('No data')
```
