# Tower of Hanoi - Experiment Analysis
Loads per-step and summary CSVs from the most recent run and plots key metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Find the most recent CSVs
steps_files = sorted(glob.glob("../output/experiments/*_steps.csv"))
summary_files = sorted(glob.glob("../output/experiments/*_summary.csv"))

assert steps_files, "No *_steps.csv found in ../output/experiments/"
assert summary_files, "No *_summary.csv found in ../output/experiments/"

csv_path = steps_files[-1]
print(f"Loading steps: {csv_path}")
df = pd.read_csv(csv_path)

summary = pd.read_csv(summary_files[-1])
print(f"Loading summary: {summary_files[-1]}")
display(summary)
```

## Fallback Retries per Step

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(df["step"], df["fallback_retries"], color="coral")
ax.set_xlabel("Step")
ax.set_ylabel("Fallback Retries")
ax.set_title("Fallback Retries per Step")
ax.set_xticks(df["step"])
plt.tight_layout()
plt.show()
```

## Cumulative Failed Predictions

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df["step"], df["num_failed_predictions"].cumsum(), marker="o", color="tomato")
ax.set_xlabel("Step")
ax.set_ylabel("Cumulative Failed Predictions")
ax.set_title("Cumulative Agent Failures Over Time")
ax.set_xticks(df["step"])
plt.tight_layout()
plt.show()
```

## Agents Queried per Step

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(df["step"], df["num_agents_queried"], color="steelblue")
ax.set_xlabel("Step")
ax.set_ylabel("Agents Queried")
ax.set_title("Number of Agents Queried per Step")
ax.set_xticks(df["step"])
plt.tight_layout()
plt.show()
```

## Consensus Rate & Summary Stats

```python
consensus_rate = df["consensus_reached"].mean() * 100
print(f"Consensus rate: {consensus_rate:.1f}%")
print(f"Total steps: {len(df)}")
print(f"Steps with fallback: {(df['fallback_retries'] > 0).sum()}")
print(f"Steps with cycle hit: {df['cycle_hit'].sum()}")
print(f"Average agents per step: {df['num_agents_queried'].mean():.1f}")
print(f"Total failed predictions: {df['num_failed_predictions'].sum()}")

fig, ax = plt.subplots(figsize=(6, 4))
consensus_yes = df["consensus_reached"].sum()
consensus_no = (~df["consensus_reached"]).sum()
ax.pie([consensus_yes, consensus_no],
       labels=["Consensus", "No Consensus"],
       autopct="%1.1f%%",
       colors=["#66bb6a", "#ef5350"])
ax.set_title("Consensus Rate")
plt.tight_layout()
plt.show()
```
